## 12. 获取行、列的平均值

```python
df = pd.DataFrame([[10, 20, 30, 40], [7, 14, 21, 28], [5, 5, 0, 0]],
                  columns=['Apple', 'Orange', 'Banana', 'Pear'],
                  index=['Basket1', 'Basket2', 'Basket3'])
display(df)
```

```python
# 1、按列统计均值
df.mean(axis=0)
```

```python
# 2、按行统计均值
df.mean(axis=1)
```

```python
# 3、增加一列'Basket mean'，该列为行均值
df['Basket mean']=df.mean(axis=1)
df
```

```python
# 4、增加一行'Fruit mean'，该行为列均值
df.loc['Fruit mean']=df.mean(axis=0)
df
```

## 16. map()、applymap()使用

```python
data = {'fruit':['apple','banana','grape'],'price':[5,6,7]}
df = pd.DataFrame(data)
display(df)
```

```python
# 1、编写函数将price中整数改为字串，例如5改为5元
df['price'].map(lambda x: str(x)+'元')
```

```python
# 2、将fruit列中首字母大写
df['fruit']=df['fruit'].str.title()
df
```

```python
df = pd.DataFrame(np.random.rand(3,3), index=['a','b','c'], columns=['A','B','C'])
display(df)
```

```python
# 将所有数据改为小数点后三位，采用applymap()
df.applymap(lambda x:round(x,3))
```

## 17. 采用“豆瓣电影数据”进行分析

```python
# 2、显示后5行
df.iloc[5:,:]
```
