# 2. Pandas 分组统计

```python
data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(data)
df
```

```python
#### 将列名修改为中文，regiment为 团，company为 连，	name为 名字，preTestScore 为 测试前分数，postTestScore为 测试后分数。
df.rename(columns={'regiment':'团','company':'连','name':'名字','preTestScore':'测试前分数','postTestScore':'测试后分数'},inplace=True)
df
```

```python
#### 计算每个 团 的 测试前分数 均值。
df.groupby('团')['测试前分数'].mean()
```

```python
#### 按 团 连 分组，计算 测试后分数 的 均值, 不以分组键为索引。（结果是DataFrame，不是Series）
df.groupby(['团','连'], as_index=False)['测试后分数'].mean()
```

```python
#### 按 团 连 分组，计算 测试前分数 的 最小值 和测试后分数 的 最大值。
df.groupby(['团','连']).agg({'测试前分数':min,'测试后分数':max})
```

# 4. Pandas数据分组

```python
# 加载数据
df = pd.read_csv('u.user', sep = '|', index_col='user_id') # 职业数据
display(df.info())
df.head()
```

```python
## 问题1：计算每种职业的平均年龄。
df.groupby('occupation')['age'].mean()
```

```python
## 问题2：计算每种职业的男性占比，并从大到小排序。（难，选做）
(df[df['gender'] == 'M']['occupation'].value_counts() / df['occupation'].value_counts()* 100).sort_values(ascending=False)
```

```python
## 问题3： 对于每种职业，计算最小和最大年龄。
df.groupby('occupation')['age'].agg({min,max})
```

```python
## 问题4：按职业、性别分组，计算平均年龄。 
df.groupby(['occupation','gender'])['age'].mean()
```

```python
## 问题5：对于每种职业，显示男女性别占比。（难，选做）
(df.groupby(['occupation', 'gender']).size() / df.groupby('occupation').size() * 100).unstack()
```

# 5. Pandas函数映射

```python
# 准备数据
df = pd.read_csv('student-mat.csv') #美国学生饮酒消费数据 
display(df.info())
df.head()
```

```python
## 问题1：切片列school到列guardian。
df.loc[:,'school':'guardian']
```

```python
## 问题2：创建lambda函数用于将字符串首字母大写，并应用到Mjob和Fjob列数据。
df[['Mjob','Fjob']].agg(lambda x:x.str.title())
```

```python
## 问题3：创建函数is_legal()，并应用到数据集，返回一个新列legal_drinker，列元素值为布尔类型，True表示学生年龄大于等于17（合法饮酒），False表示小于17（不合法饮酒）。
def is_legal(age):
    return age >= 17
df['legal_drinker'] = df['age'].apply(is_legal)
df
```

# 6. Pandas 数据合并和连接

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df1 = pd.DataFrame(raw_data_1)

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df2 = pd.DataFrame(raw_data_2)

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df3 = pd.DataFrame(raw_data_3)
print(df1)
print(df2)
print(df3)
```

```python
#### 将df1和df2按行合并为df_row，并重新索引。
df_row = pd.concat([df1,df2],ignore_index=True)
df_row
```

```python
#### 将df1和df2按列合并为df_col。
df_col = pd.concat([df1,df2],axis=1)
df_col
```

```python
#### 将df_row和df3按照 subject_id 进行内连接。
pd.merge(df_row, df3, on='subject_id', how='inner')
```

```python
####  将具有相同 subject_id 值的df1和df2连接。
pd.merge(df1, df2, on='subject_id')
```

```python
#### 将df1和df2的所有值按 subject_id 合并连接（外连接）。
pd.merge(df_row, df3, on='subject_id', how='outer')
```

# 7. Pandas 缺失值处理

```python
df = pd.read_csv('property-data.csv')
df
```

```python
#### 删除含有空字段的行。
df.dropna()
```

```python
#### 用0代替所有的空字段。
df.fillna(0)
```

```python
#### 删除一行具有两个以及两个以上缺失值的行。
df.dropna(thresh=6)
```

```python
#### 删除一列具有两个以及两个以上缺失值的列。
df.dropna(thresh=len(df)-1,axis=1)
```

```python
#### ST_NUM 列的缺失值用均值代替。
df['ST_NUM'].fillna(df['ST_NUM'].mean(),inplace=True)
df
```

```python
#### NUM_BEDROOMS 列的缺失值使用前向填充方式替代
df['NUM_BEDROOMS'].fillna(method='ffill', inplace=True)
df
```
