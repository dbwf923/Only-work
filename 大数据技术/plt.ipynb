```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
import pandas as pd
import seaborn as sns

text=pd.read_csv("news.txt",header=None)

words=' '.join(jieba.cut(str(text)))
print(words)

stop_word=("en_stopwords.txt")
final=''
for word in words:
    if word not in stop_word:
        final += word
print(final)

wordcloud = WordCloud(font_path='微软雅黑.ttf', width=800, height=400, background_color='white').generate(words)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()


#热力图
data = pd.read_csv("EData.csv")

plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, cmap='viridis')
plt.title('Heatmap of CSV Data')
plt.show()
```
